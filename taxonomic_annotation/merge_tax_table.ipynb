```python
import pandas as pd
from functools import reduce
from pathlib import Path
import os
```

```python
gene_name="cNOR"
tax_level="family"
```

```python
tax_list = [ f for f in os.listdir('.') if f.endswith(tax_level)]
```

```python
tax_df_list = []
for file in tax_list:
    clean_df = "df_"+file
    clean_df = pd.read_csv(file, sep="\t")
    tax_df_list.append(clean_df)
```

```python
all_combined = reduce(lambda  left,right: pd.merge(left,right,on=tax_level,how='outer'), tax_df_list)
```

```python
all_combined = all_combined.reindex(sorted(all_combined.columns), axis=1)
all_combined = all_combined[ [tax_level] + [ col for col in all_combined.columns if col != tax_level ] ]
all_combined
```

```python
pd.DataFrame.to_csv(all_combined, gene_name+'___'+tax_level+'___Black_sea_MG.tsv', sep='\t', index=False)
```
